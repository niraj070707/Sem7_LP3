```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function and its derivative
def function(x):
    return (x + 3) ** 2

def gradient(x):
    return 2 * (x + 3)
```

```python
# Gradient Descent Algorithm
def gradient_descent(starting_point, learning_rate, num_iterations):
    x = starting_point
    x_values = [x]  # Store x values for plotting

    for i in range(num_iterations):
        grad = gradient(x)  # Calculate the gradient
        x = x - learning_rate * grad  # Update x
        x_values.append(x)  # Store the updated x value
        print(f"Iteration {i + 1}: x = {x}, f(x) = {function(x)}")  # Print the current state
        
    return x, x_values
```

```python

# Parameters
initial_x = 2
learning_rate = 0.1
iterations = 10

# Run the Gradient Descent Algorithm
local_minima, x_values = gradient_descent(initial_x, learning_rate, iterations)
```

```python
x_range = np.linspace(-6, 2, 200)
y_range = function(x_range)

plt.figure(figsize=(10, 6))
plt.plot(x_range, y_range, label='y = (x + 3)²', color='blue')
plt.scatter(x_values, function(np.array(x_values)), color='red', label='Gradient Descent Path', zorder=5)
plt.title('Gradient Descent on the Function y = (x + 3)²')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()
plt.show()
```
